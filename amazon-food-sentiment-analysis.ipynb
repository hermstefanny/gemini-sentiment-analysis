```python
import kaggle
import pandas as pd
from time import sleep
import random
```

```python
kaggle.api.authenticate()
kaggle.api.dataset_download_files("snap/amazon-fine-food-reviews", path = '.', unzip=True)
```

```python
reviews_df = pd.read_csv("Reviews.csv")
reviews_df
```

```python
reviews_summ_text = reviews_df[["Summary","Text"]].sample(n=10, random_state=5)
reviews_summ_text
```

### Gemini

```python
from dotenv import load_dotenv
import os
from google import genai

# Load environment variables from .env file
load_dotenv()

# Get the API key from the environment
API_KEY = os.getenv("GOOGLE_API_KEY")

# Check if the API key is loaded properly
if not API_KEY:
    raise Exception('GOOGLE_API_KEY not found. Please set it in your .env file.')

# Use the API key
client = genai.Client(api_key=API_KEY)
```

```python
gemini_response =list()

google_chosen_model='gemini-2.5-pro'
for index, review in reviews_summ_text.iterrows():
    response = client.models.generate_content(
        model=google_chosen_model, 
        contents=f"""You are a NLP helper for a company that investigates customer's satisfaction about its products.
        Perform a sentiment analysis review on this text {review['Text']} 
        Give your insights in this structured manner:
        'Overall Sentiment' (one to three words): 
        'Explanation for this sentiment':
        'Key insigths for the company':
        """
    )

    gemini_response.append(response.text)
    print(f"Response for review {index} acquired")
    print(f"Response feedback {response.prompt_feedback}")
    sleep(random.randint(2,15))

print("Proccess finished")   
```

```python
gemini_response
```

```python
reviews_summ_text['GeminiEval'] = gemini_response
```

```python
reviews_summ_text
```

### ChatGPT


```python
from openai import OpenAI
from dotenv import load_dotenv
import tiktoken
```

```python
gpt_response =list()
## LOADING API KEY FROM THE .env FILE
load_dotenv()
key_ct = os.environ["ct_api_key"]
client = OpenAI(api_key = key_ct)

encoding = tiktoken.encoding_for_model("gpt-4o-mini")

for index, review in reviews_summ_text.iterrows():

    instruction = f"""You are a NLP helper for a company that investigates customer's satisfaction about its products.
        Perform a sentiment analysis review on this text {review['Text']} 
        Give your insights in this structured manner:
        'Overall Sentiment' (one to three words): 
        'Explanation for this sentiment':
        'Key insigths for the company':
        """ 


    response = client.chat.completions.create(
        model = "gpt-4o-mini",
        messages = [
            {"role": "system", "content" : instruction},
        ],
        temperature = 0.80,
        max_tokens = 500,
    )
   
    gpt_response.append(response.choices[0].message.content)
    print(f"Response for review {index} acquired")
    sleep(random.randint(2,6))

print("Process for GPT 4.o mini finished")
```

```python
gpt_response
```

```python
reviews_summ_text['GPTEval'] = gpt_response
```
